```
from matplotlib import pyplot as plt
import numpy as np
from matplotlib import style

style.use("ggplot")
x = np.arange(1,1000)
y = x*x;
plt.plot(x,y, 'g', label="exp function", linewidth=2)

plt.title("first demo")
plt.ylabel("Y axis")
plt.xlabel("X axis")

x = np.arange(1,100)
y = x*x*x;
plt.plot(x,y, 'r', label="cube function", linewidth=5)

plt.title("first demo")
plt.ylabel("Y axis")
plt.xlabel("X axis")

plt.legend()
plt.show()

def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
abline(100,10)
```
