```
import numpy as np
import matplotlib.pyplot as plt
H = np.array([[1, 0, 0, 4],[1, 2, 1, 4]])
#fig = pmt.figure(figsize=(6,3.2))
plt.imshow(H)
plt.show()

x = np.arange(1, 11)
print("Plotting exp data")
print(x);  
y = x * x 
print(y);
# plotting 
plt.title("Line graph")  
plt.xlabel("X axis")  
plt.ylabel("Y axis")  
plt.plot(x, y, color ="red")  
plt.show()

x = np.arange(1, 11)  
y = np.array([100, 10, 300, 20, 500, 60, 700, 80, 900, 100]) 

# plotting 
plt.title("Line graph")  
plt.xlabel("X axis")  
plt.ylabel("Y axis")  
plt.plot(x, y, color ="green")  
plt.show()

V = np.array([[1,1], [2,-2], [4,-7]])
origin = np.array([[0, 0, 0],[0, 0, 0]]) # origin point

plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], scale=21)
plt.show()

V = np.array([[5,1], [1,5]])
origin = np.array([[0, 0],[0, 0]]) # origin point

plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], scale=20)
plt.show()
print(V[:,0])
print(V[:,1])
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'],  angles='xy', scale_units='xy', scale=1)
v12 = V[0] + V[1] # adding up the 1st (red) and 2nd (blue) vectors
print(v12)
plt.quiver(*origin, v12[0], v12[1],  angles='xy', scale_units='xy', scale=1)
plt.show()
```
